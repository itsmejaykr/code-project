```python
import pandas as pd
import re
import csv
from bs4 import BeautifulSoup
import requests as req
```

```python
url = "https://engineering.careers360.com/colleges/ranking?page=1&icn=engineering-ranking_landing-page&ici=data-section_c360-home&sort_by=1&stream=1"
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36",
    "Referer": url,
}
try:
    res = req.get(url,headers=headers)
    if res.status_code == 200:
        print("Access_Granted!",res.status_code)
    else:
        print("Access_Deined",res.status_code)
except requests.exceptions.RequestException as e:
    print("Error!",e)
```

```python
def extract_rank(text):
    match = re.search(r"(\d{1,3}-\d{1,3}|(\d+))\s*NIRF Rank'\d{2}", text)
    if match:
        rank = match.group(1)
        return rank
    else:
        return None 
def extract_fees(text):
    pattern = r'\d+(?:\.\d{2})?(?: Lakhs| K)'
    matches = re.findall(pattern, text)
    if matches:
        for match in matches:
            if 'Lakhs' in match:
                amount = float(match.replace(' Lakhs', '')) * 100000
                return int(amount)
            elif 'K' in match:
                amount = float(match.replace(' K', '')) * 1000
                return int(amount)
            else:
                return 0

```

```python
college_name = []
Own = []
Rank = []
Courses = []
Fees = []
for i in range(1,3):
    url = "https://engineering.careers360.com/colleges/ranking?page{0}&icn=engineering-ranking_landing-page&ici=data-section_c360-home&sort_by=1&degree=2&state=12&stream=1".format(i)
    headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36",
    "Referer": url,}
    response = req.get(url,headers = headers)
    soup = BeautifulSoup(response.content,"html.parser")
    collage_name = soup.find_all("h3",class_ = "college_name d-md-none")
    O = soup.find_all("div" , class_="content_block d-block d-md-none")
    Ranks = soup.find_all("div",class_ = "align-center")
    ul = soup.find_all("ul",class_ = "snippet_list")
    
    for college,ownership,ranks,course in zip(collage_name,O,Ranks,ul):
        college_name.append(college.text.strip())  
        Own.append(ownership.text.strip())
        #Rank.append(extract_rank(ranks.text.strip()))
        Rank.append(extract_rank(ranks.text.strip()))
        Courses.append(course.text.strip())
        Fees.append(extract_fees(course.text.strip()))
        
    Data = pd.DataFrame({"NIRF_Rank":Rank,"College": college_name,"Ownership":Own,"Courses":Courses,"FEE":Fees})
    
        
```
